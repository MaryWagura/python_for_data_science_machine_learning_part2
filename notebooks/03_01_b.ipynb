## K-means method
### Setting up for clustering analysis

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

import sklearn
from sklearn.preprocessing import scale
import sklearn.metrics as sm
from sklearn.metrics import confusion_matrix, classification_report
```

```python
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
```

```python
%matplotlib inline
plt.figure(figsize=(7,4))
```

```python
iris_dataset = datasets.load_iris()

x = scale(iris_dataset.data)
y = pd.DataFrame(iris_dataset.target)

names= iris_dataset.feature_names
x[0:10]
```

## Building and running your model

```python
clustering = KMeans(n_clusters=3, random_state=5).fit(x)
```

## Plotting your model outputs


```python
iris_df = pd.DataFrame(iris_dataset.data)
iris_df.columns = [
    'Sepal_Length',
    'Sepal_Width',
    'Petal_Length',
    'Petal_Width']
y.columns = ['Targets']
```

```python
color_theme = np.array(['red', 'lightsalmon', 'powderblue'])

plt.subplot(1,2,1)

plt.scatter(
    x=iris_df.Petal_Length,
    y=iris_df.Petal_Width,
    color=color_theme[iris_dataset.target],)
plt.title('Ground Truth Classifier')

plt.subplot(1,2,2)

plt.scatter(
    x=iris_df.Petal_Length,
    y=iris_df.Petal_Width,
    color=color_theme[clustering.labels_],)
plt.title('K-Means Clustering')
```
